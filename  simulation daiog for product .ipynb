```python
# Message Body 
offer_data.head()
def offering(offer_data, prodect):
    descriptions = [] 

    result = offer_data[offer_data['التصنيف'] == prodect]

    for index, row in result.iterrows():
        name        = row['اسم المنتج']
        description = row['المواصفات']
        quantity    = row['الكمية'] 
        price       = row['السعر'] 
        p_curr      = row['العملة']

        quantity_massge = 'وهو متوفر حاليا' if quantity != 0 else 'ولكن للاسف نفذت الكمية وسنعمل على طلبها وتوفرها لدينا باسرع ما يمكن '
        price_curr = str(price)+ " " + p_curr
        
        adaptation = random.choice([1, 2, 3])
        if adaptation == 1 or adaptation == 2: 
            item_description = f'- {name}: {description}, {quantity_massge}'
        else:
            item_description = f'- {name}: {description}, {quantity_massge}, وتبلغ تكلفتة {price_curr}'
        descriptions.append(item_description)
    return descriptions


def message_offering(offer_data):
    def message_offering_prodect(prodect):
        offering_p = offering(offer_data, prodect)

        start_offering     = '\n لدينا عرض ل[تصنيف] متاحة للبيع. سازودكم  ببعض الموديلات الشهيرة هي'
        formatted_offering = f' \n {offering_p[0]} \n {offering_p[1]} \n'
        end_offering       = 'ارجوا ان تخبرنا اي عرض نال اعجابكم'

        message = start_offering + formatted_offering + end_offering
        
        return message
#     message.replace('[تصنيف]', prodect)
#     print(message)
    return message_offering_prodect
```

```python
def message_description(offer_data, new= 0):
    def message_description_prodect(prodect, adaptation = 8):
        descriptions_prodect = []
        result = offer_data[offer_data['اسم المنتج'] == prodect]
        for index, row in result.iterrows():
            name        = row['اسم المنتج']
            description = row['المواصفات']
            quantity    = row['الكمية'] 
            price       = row['السعر'] 
            p_curr      = row['العملة']
    
            quantity_massge = 'وهو متوفر حاليا' if quantity != 0 else 'ولكن للاسف نفذت الكمية وسنعمل على طلبها وتوفرها لدينا باسرع ما يمكن '
            price_curr = str(price)+ " " + p_curr
        
#             item_description = f'- {name}: {description}, {quantity_massge}, وتبلغ تكلفتة {price_curr}'
#             descriptions_prodect.append(item_description)
           
            adaptation = random.choice([1, 2, 3, 4, 5, 6, 7, 8])
    
            if new != 0:
                adaptation = new
    
            if   adaptation == 1:
                descriptions_prodect.append(f'- {name}: {description}')
            elif adaptation == 2:
                descriptions_prodect.append(f'- {name}: {quantity_massge}')
            elif adaptation == 3:
                descriptions_prodect.append(f'- {name}: وتبلغ تكلفتة {price_curr}')
            elif adaptation == 4:
                descriptions_prodect.append(f'- {name}: {description}, {quantity_massge}')
            elif adaptation == 5:
                descriptions_prodect.append(f'- {name}: {description}, وتبلغ تكلفتة {price_curr}')
            elif adaptation == 6:
                descriptions_prodect.append(f'- {name}: {quantity_massge}, وتبلغ تكلفتة {price_curr}')
            elif adaptation == 7:
                descriptions_prodect.append(f'- {name}: {quantity_massge}, {description}, وتبلغ تكلفتة {price_curr}')
            elif adaptation == 8:
                descriptions_prodect.append(f'- {name}: {description}, {quantity_massge}, وتبلغ تكلفتة {price_curr}')
                
                
    
        start_offering     = '\n اليك تفاصيل [المنتج] وهي '
        formatted_offering = f' \n {descriptions_prodect[0]}'

        message = start_offering + formatted_offering
        
        return message
    return message_description_prodect
```

```python
def pordect_guaranteed(offer_data):
    def guaranteed_prodect(prodect):
        guaranteed = offer_data[offer_data['اسم المنتج'] == prodect]
        my_list = guaranteed['المواصفات'].str.split().tolist()[0]
        if 'مكفول' in my_list:
            # Find the index of 'مكفول' in the list
            index = my_list.index('مكفول')
            result = ' '.join(my_list[index:index+3])
        else:
            result = 'مش مكفول '
        return result
    return guaranteed_prodect
```

```python
def default_encoder(obj):
    if isinstance(obj, np.int64):
        return int(obj)
    raise TypeError(f"Object of type {obj.__class__.__name__} is not JSON serializable")


# Convert datafram to JSON
def save_json(df, name_file = './arabic_conversation_simulation.json'):
    dialog_data = []
    for dialog in df['id_dialog'].unique():
        tmp ={
        "dialog_id": dialog,
        "dialog": []
         }
        count = 0
        result = df[df['id_dialog'] == dialog]

        for index, row in result.iterrows():    
            if row['massage'] != '':
                tmp["dialog"].append({"id": count, "sender": 'participant2', "text": row['massage']}) # participant2 --> custmer 
                count +=1
            tmp["dialog"].append({"id": count, "sender": 'participant1', "text": row['replies']}) # participant1 --> company  
            count +=1
        dialog_data.append(tmp)
        
    json_object = json.dumps(dialog_data, ensure_ascii=False, indent=4, default=default_encoder)
    # Save JSON data to a file
    with open(name_file, "w", encoding="utf-8") as json_file:
        json_file.write(json_object)

    print("JSON data saved to data.json")

#     return dialog_data
save_json(df)
# evaluation genaration dialogue model
```
